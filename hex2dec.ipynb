```python
# Function input: text file with strings (Hexadecimal values)
# Function output: array of corresponding decimal values
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read from file and print
text_file = open("my_file.txt", "r")
lines = text_file.read().splitlines()
#print (lines)
```

```python
# converting hexadecimal string to decimal 
def hex_str_2_dec(my_list = []):
    result = []
    for x in my_list:
        result.append(int(x,16))
    return result 
```

```python
# Convert hexadecimal to decimal
accln_code = hex_str_2_dec(lines)
```

```python
# Find physical value
accln_array = np.array(accln_code)
accln = (0.002 * accln_array.astype(float)) - 65.000
```

```python
# plot data
plt.figure()
plt.plot(accln)

# mark 6 m/s^2
x = np.linspace(0,550,100)
y = 6 * np.ones(len(x))
z = -y
ref = np.zeros(len(x))

plt.plot(x, y, '-r', label='accln = 6 m/s^2')
plt.plot(x, z, '-g')
plt.plot(x, ref, '-k', label='accln = -6 m/s^2')

# labels
plt.title('Longitudinal Acceleration')
plt.xlabel('Index')
plt.ylabel('Long_Accln (m/s^2)')
plt.axis((0,550,-65,10))
```
